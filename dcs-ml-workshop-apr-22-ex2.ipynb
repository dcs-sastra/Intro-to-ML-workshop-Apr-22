# `Confusion Matrix`
<h4>True Positives (TP) :<\h4> <h5>Model predicted 1 and actual value is also 1<br>
<h4>True Negatives (TN) :<\h4> <h5>Model predicted 0 and actual value is also 0.<br>
<h4>False Positives (FP) :<\h4> <h5>Model predicted 1 but actual value is 0. (Also known as a "Type I error.")<br>
<h4>False Negatives (FN) :<\h4> <h5>Model predicted 0 but actual value is 1. (Also known as a "Type II error.")<br><br>

# `Classification Report`
<h4>Precision is the accuracy of positive predictions.<br>Precision = TP/(TP + FP)
<br>
<h4>Recall is the ability of a classifier to find all positive instances.<br>Recall = TP/(TP + FN)
<br>
<h4>F1 score conveys the balance between the precision and the recall.<br>F1 Score = 2 * ( ( precision *  recall ) / ( precision + recall ) )

```python
cm = confusion_matrix(y_test_clf, lr_preds, labels = lr_clf.classes_)

disp = ConfusionMatrixDisplay(
    confusion_matrix = cm,
    display_labels = lr_clf.classes_
)
disp.plot()
plt.show()

print(classification_report(y_test_clf, lr_preds))
```
